# Finding info in the wine data set

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.show()
sns.set()
```

```python
dfRed = pd.read_csv('winequality-red.csv', sep=';')
dfWhite = pd.read_csv('winequality-white.csv', sep=';')
```

```python
dfRed.head()
```

```python
dfWhite.head()
```

```python
sns.jointplot(x = 'alcohol', y = 'quality', data = dfRed, kind = 'reg')
```

```python
sns.jointplot(x = 'alcohol', y = 'quality', data = dfWhite, kind = 'reg')
```

```python
sns.pairplot(dfRed, hue = 'quality', palette = 'coolwarm')
# Not very informative
```

```python
sns.violinplot(x = 'quality', y = 'alcohol', data = dfRed, dodge = True)
```

```python
dfRed['alcohol'].mean()
```

```python
dfWhite['alcohol'].mean()
# White wine has a higher expected alcohol content
```

```python
dfRed['alcohol'].var()
```

```python
dfWhite['alcohol'].var()
# White wine has a higher variance in alcohol content
```
